```python
import pandas as pd
import pdtypes as pdt
```

```python
pdt.unpatch()

df = pd.DataFrame({
    "x": [1, 1, 2, 2],
    "y": [1, 2, 3, 4],
})

df
```

```python
pdt.patch()
df
```

```python
pdt.unpatch()

gf = df.groupby(["x"])

gf
```

```python
pdt.patch()

gf
```

```python
pdt.unpatch()

df2 = pd.DataFrame(
    {
        "a" : [1, 2, 3, 4, 5, 6],
        "b": [7, 8, 9, 10, 11, 12],
        "c": [df] * 6,
    }
)
df2
```

```python
pdt.patch()

df2
```

```python
# hierarchically indexed

df3 = df2.set_index(["a", "b"])
df3.columns = pd.MultiIndex.from_tuples([("c", "c1")])
df3
```
